```python
# Given an integer partition mu, return the 
# corresponding linear combination of simple
# roots
def weightify (mu, r):
    out = [sum(mu[0:i + 1]) for i in range(len(mu))]
    # Pad out the rest
    out = out + [out[-1]] * (r - len(mu));
    
    return out;
```
